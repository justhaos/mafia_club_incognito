```python
import pandas as pd
import numpy as np
from lxml import etree
from bs4 import BeautifulSoup as BS
import requests
```

```python
main_url = "https://gomafia.pro"
sort_by_elo = "/rating?sortTypeUsers=elo&sortOrderUsers=desc&regionUsers={}&pageUsers={}"

session = requests.Session()
df_data = []

regions = ['central', 'chernozem_region', 'north', 'south', 'volga_region', 'siberia_and_ural']
for r in regions:
    min_elo = 4000
    i = 0
    while min_elo >= 2200:
        i += 1
        html_content = session.get(main_url+sort_by_elo.format(r, i)).content
        soup = BS(html_content, "html.parser")
        table = soup.find("table")
        for row in table.find('tbody').find_all('tr'):
            data_row = row.find_all('td')
            nick = data_row[1].find('a').text
            link = main_url+data_row[1].find('a')['href']
            club = data_row[1].find_all('span')[-1].text
            gg = int(data_row[3].text)
            elo = int(data_row[4].text)
            min_elo = min(min_elo, elo)
            df_row = [r, link, nick, club, gg, elo]
            df_data.append(df_row)

df = pd.DataFrame(df_data, columns=['region', 'link', 'nick', 'club', 'gg', 'elo'])
df
```
